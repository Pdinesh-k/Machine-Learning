#1) Splitting the data into train,test,val

let's experiment with different depth using a helper function

```python
def max_depth_error(md):
  model = DecisionTreeClassifier(random_state=42,max_depth = md)
  model.fit(x_train,train_target)
  val_error = 1- (model.score(x_val,val_target))
  train_error = 1- (model.score(x_train,train_target))
  return{ "max_depth" : md , "val_error" : val_error , "train_error" : train_error}
```

```python
error_df = pd.DataFrame([max_depth_error(md) for md in range(1,21)])
```

```python
error_df
```

```python
plt.plot(error_df["max_depth"],error_df["train_error"])
plt.plot(error_df["max_depth"],error_df["val_error"])
plt.title("Training vs Validation Error")
plt.xlabel("Max_Depth")
plt.ylabel("Prediction error")
plt.legend(["Training","validation"])
```

In this our graph we can identify that at depth 7 it has the highest accuracy for validation

so we can look at depth 7 how much is the accuracy is

```python
model = DecisionTreeClassifier(random_state=42,max_depth=7).fit(x_train,train_target)
model.score(x_train,train_target) , model.score(x_val,val_target)
```

#Training a Random Forest

While tuning the hyperparameters of a single decision tree may lead to some improvements , a much more effective strategy is to combine the results of several decision trees trained with slightly different parameters . This is called a random forest model.

The key idea here is that each decision tree in the forest will make different kinds of errors and upon averaging many of their errors will cancel out . This idea is commonly known as the "wisdom of the crowd"

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs = -1 , random_state = 42)
```

```python
model.fit(x_train,train_target)
```

Noe we can give the value for parameters max_depth , n_estimators , max_leaf_nodes

```python
def try_parameter(**params):
  model = RandomForestClassifier(random_state = 42 , n_jobs = -1 , **params).fit(x_train,train_target)
  return{model.score(x_train,train_target) , model.score(x_val,val_target)}
```

```python
try_parameter(max_depth=5 , max_leaf_nodes = 720 , n_estimators = 7)
```

```python
try_parameter(max_depth=10 , max_leaf_nodes = 1400 , n_estimators = 10)
```

```python
try_parameter(max_leaf_nodes = 2**5)
```

```python
try_parameter(max_leaf_nodes = 2**20)
```

```python
try_parameter(max_leaf_nodes = 2**25)
```

#min_sample_split and min_samples_leaf

By default , the decision tree classifier tries to split every node that has 2 or more . You can increase the values of these arguments to change this behaviour and reduce overfitting , especially for very large dataset

```python
try_parameter(min_samples_split=5 , min_samples_leaf=2)
```

The tree will be splitted if there are minimum of 5 rows reach and leaf nodes with minimum of 2

```python
try_parameter(min_samples_split=10 , min_samples_leaf = 5)
```

#min_impurity_decrease

A node will be split if this split induces a decrease of the impurity(Gini Index) greater than or equal to this value . It's default value is 0 , and you can increase it to reduce overfitting.

```python
try_parameter(min_impurity_decrease = 1e-6)
```

```python
try_parameter(min_impurity_decrease = 1e-4)
```

```python
try_parameter(min_impurity_decrease = 1e-2)
```

```python
try_parameter(min_impurity_decrease = 1e-1)
```

#Bootstrap :

In RandomForest bootsrap is the process in which the model takes the one row and take value from it and what it had taken it will be replaced again for other Decision Tree

By default in Random Forest Bootstrap is enabled always.

We can make it Bootstrap as False and we can check the result

```python
try_parameter(bootstrap = False)
```

If Bootsrap is set to False the one row will be taken by every Decision Tree

```python
try_parameter(bootstrap = True)
```

When bootstrapping is enabled you can control the number of fraction of rows to be considered for each bootsrap using max_sample

```python
try_parameter(max_samples = 0.5)
```

```python
try_parameter(max_samples = 0.9)
```
